```python

import pandas as pd

df = pd.read_csv("yearsclean.csv", index_col="PKid")
```

1. Calculo de duracion de proyectos

```python
ultimo_ano = df.groupby('id')['year'].max()
ultimo_ano
```

```python
duracion = ultimo_ano - df.groupby('id')['fecha_alta'].min()
duracion
```

```python
duracion.to_csv("duraciones.csv")
```

```python
duracion_df = pd.DataFrame({'id': duracion.index, 'duracion': duracion.values})
```

```python
dfduracion = pd.merge(df, duracion_df, on='id')
```

```python
dfduracion.to_csv("datasetFinalNL.csv")
```

```python
#USANDO AGRUPACIONES DE CLUSTERIZACION PARA REDUCIR DIMENSIONALIDAD
data = dflimpio[:]
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Imagina que 'datos' es tu DataFrame con una columna 'giro_negocio'
datos_limpieza = data['nombre_act'].str.lower().str.replace('[^\w\s]','')

# Tokenización y eliminación de palabras vacías
stop_words = stopwords.words('spanish')
datos_limpieza = datos_limpieza.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))

# TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(datos_limpieza)

# K-means Clustering
num_clusters = 100  # Ajusta este número según tus necesidades
km = KMeans(n_clusters=num_clusters)
km.fit(X)

# Obtener los clusters
clusters = km.labels_.tolist()
```

```python
dflimpio["giro"] = clusters
```

```python
dflimpio.to_csv("giroclusterFINN.csv")
```

```python
x = dfxencoded.drop("duracion", axis=1)
y = dfxencoded[["duracion"]]

dfxencoded.drop("duracion", axis=1)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 2. Cargar o crear un conjunto de datos
# Por ejemplo, vamos a usar el conjunto de datos de Iris incluido en scikit-learn
from sklearn.datasets import load_iris

X = x  # características
y = y.values.ravel()  # etiquetas

# 3. Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=54)

# 4. Crear el clasificador KNN y ajustarlo a los datos de entrenamiento
knn = KNeighborsClassifier(n_neighbors=5)  # Usar 3 vecinos para KNN
knn.fit(X_train, y_train)

# 5. Predecir las etiquetas para el conjunto de datos de prueba
y_pred = knn.predict(X_test)

# 6. Calcular y mostrar la precisión (accuracy) del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy del modelo KNN: {accuracy:.2f}")
```
