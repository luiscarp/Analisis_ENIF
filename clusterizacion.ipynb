```python
import pandas as pd
```

```python
data = pd.read_csv("dfCombined (2).csv")

data.info(
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Imagina que 'datos' es tu DataFrame con una columna 'giro_negocio'
datos_limpieza = data['nombre_act'].str.lower().str.replace('[^\w\s]','')

# Tokenización y eliminación de palabras vacías
stop_words = stopwords.words('spanish')
datos_limpieza = datos_limpieza.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))

# TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(datos_limpieza)

# K-means Clustering
num_clusters = 50  # Ajusta este número según tus necesidades
km = KMeans(n_clusters=num_clusters)
km.fit(X)

# Obtener los clusters
clusters = km.labels_.tolist()
```

```python
clusters
```

```python
# Agregar la lista de clusters como una nueva columna en el DataFrame
data['cluster'] = clusters

# Opcionalmente, puedes ordenar el DataFrame basado en los clusters
datos_ordenados = data.sort_values('cluster')
```
